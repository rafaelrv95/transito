```python
cd = covid.groupby(['departamento_nom'])[['edad']].describe()
cd
```

```python
cd.columns
```

```python
cd.columns = cd.columns.droplevel(0)
```

```python
cd['count']
```

```python
#graficar la contidad de casos por departamento
fig, ax = plt.subplots(figsize=(16,6))
ax.bar(x=cd.index, height = cd['count'], color='magenta', alpha=0.4)
#https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation= 45, ha='right')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.bar(x=cd.index, height = cd['mean'], color='red', alpha=0.4)
#https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation= 45, ha='right')
plt.show()
```

```python
covid.columns
```

```python
#acumulado de caos por cada dia
for fecha in ['fecha_reporte_web','fecha_de_notificaci_n', 'fecha_inicio_sintomas',
       'fecha_muerte', 'fecha_diagnostico', 'fecha_recuperado']:
    covid[fecha] = pd.to_datetime(covid[fecha], format = '%d/%m/%Y %H:%M:%S')
```

```python
covid.info()
```

```python
cf = covid.groupby(['fecha_diagnostico'])[['edad']].describe()
cf
```

```python
cf.columns = cf.columns.droplevel(0)
cf
```

```python
#total de pacientes dignosticados por cada dia
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf['count'], 'b:')
ax.set(title='total de pacientes dignosticados por cada dia', xlabel='fechas', ylabel='Numero de pacientes')
#https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
plt.xticks(rotation= 45, ha='right')
plt.show()
```

```python
cf['acumulados'] = cf['count'].cumsum()
cf['acumulados']
```

```python
#total de pacientes dignosticados por cada dia
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf['acumulados'], 'g-')
ax.set(title='total de pacientes dignosticados por cada dia', xlabel='fechas', ylabel='Numero de pacientes')
#https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
#plt.xticks(rotation= 45, ha='right')
ax.grid()
plt.show()
```
